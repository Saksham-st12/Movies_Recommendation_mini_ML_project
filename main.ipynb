```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import ast
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(1)
```

```python
credits.head(1)
credits.rename(columns={'movie_id':'id'},inplace=True)
```

```python
movies = movies.merge(credits,on="id")
movies.info()
```

```python
# movies.drop(columns=['budget','homepage','original_language','original_title','popularity','production_companies','production_countries','revenue','release_date','runtime','spoken_languages','status','tagline','title_y','vote_average','vote_count'])
movies=movies[['id','title_x','genres','keywords','overview','cast','crew']]

movies.rename(columns={'id':'movie_id','title_x':'Title','cast_x':'cast','crew_x':'crew'},inplace=True)
movies.head(1)
```

```python

movies.dropna(inplace=True)
movies.iloc[0].genres
```

```python
def convert(test):
    res = []
    for i in ast.literal_eval(test):
        res.append(i['name'])
    return res
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
def top_3(TEST):
    res =[]
    count =0
    for i in ast.literal_eval(TEST):
        if count!=3:
            res.append(i['name'])
            count+=1
        else:
            break
    return res
```

```python
movies['cast']=movies['cast'].apply(top_3)
```

```python
def dir(TEST):
    res =[]
    count =0
    for i in ast.literal_eval(TEST):
        if i['job']=='Director':
            res.append(i['name'])
            break
    return res
```

```python
movies['crew']=movies['crew'].apply(dir)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.rename(columns={'crew':'Director'},inplace =True)
```

```python
movies.head(5)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","_")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","_")for i in x])
movies['Director']=movies['Director'].apply(lambda x:[i.replace(" ","_")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","_")for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['Director']
```

```python
movies
```

```python
processed_df = movies[['movie_id','Title','tags']]
processed_df
```

```python
processed_df['tags']=processed_df['tags'].apply(lambda x:" ".join(x))
processed_df
```

```python
processed_df['tags'][0]
```

```python
processed_df['tags']=processed_df['tags'].apply(lambda x: x.lower())
```

```python
processed_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
PS =PorterStemmer()
```

```python
def stem(text):
    y= []
    for i in text.split():
        y.append(PS.stem(i))
    return " ".join(y)
```

```python
processed_df['tags']= processed_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = CV.fit_transform(processed_df['tags']).toarray()
```

```python
CV.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity[4799]
```

```python
def recommend(movie):
    movie_index = processed_df[processed_df['Title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse= True,key= lambda x:x[1])[1:6]
    for i in  movies_list:
        print(processed_df.iloc[i[0]].Title)
```

```python
recommend('Batman Begins')
```
